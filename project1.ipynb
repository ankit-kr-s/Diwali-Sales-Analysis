```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
# %matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali_Sales_Data1.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
# Here we have used drop function to remove the column named das status and unnamed , 
# Axis represents which rows or columns to be removed if axis is 1 or columns it means column will be removed otherwise row.
# 
df.drop(["Status","unnamed1"], axis = "columns", inplace=True)
```

```python
# The isnull() method returns a DataFrame  where all the values are replaced with a Boolean value True for NULL values, and otherwise False.
df.isnull().sum()
```

```python
df.shape
```

```python
 # dropna function delete the null values from the data frame . and inplace() func save the changes permanentally in the file whatever we have made by using dropna func.On the other hand data will be removed but not saved , whenever we use df.info() then data will be shown
# Alternate method of not using inplace() is that we need to assign new data frame(after removing/changes in old data frame) in a new variable.
#  df.dropna(inplace=True) and new_data_frame = df.dropna() both are same.

df.dropna(inplace=True)
```

```python
df.shape
```

```python
# HERE we have check whether the null values from data fram is deleted or not. here in amount column there is 0 showing it means there is no null vales
df.isnull().sum()
```

```python
# change data type by using astype() func.
# column ko hamesa square bracket('[]')me likhte hai
df["Amount"] = df["Amount"].astype("int")
```

```python
# data type check by using dtypes.
df["Amount"].dtypes
```

```python
# To check the column we use columns
df.columns
```

```python
# we can rename column by using rename func but it is not saved in the file because we dont use inplace func. It is only showing in this cell whenever this cell is executed
df.rename(columns={"Marital_Status":"Shadi"})
```

```python
# describe() method returns description of the data in the dataframe(eg count, mean, std etc)
df.describe()
```

```python
# use describe() for specific column
df["Age"].describe()
```

```python
# if we use more than one column then we need to use double square bracket
# Square brackets [] are used to select multiple columns from a DataFrame.
# A list of column names (e.g., ["Age", "Orders", "Amount"]) is passed to the [].
df[["Age","Orders", "Amount"]].describe()
```

```python
# Exploratory Data Analysis
ax = sns.countplot(x="Gender",data = df, palette=["blue", "orange"])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
```

```python
# Here i have group by the gender and find total purchase by them and plot it on bar chart
sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(data=sales_gen, x="Gender", y="Amount")
# INSIGHTS-> from below graphs we can see that most of the buyers are females and even the purchasing power of the females are greater than men
```

```python
# find the age_group who are purchased maximum product

ax = sns.countplot(data=df, x="Age Group", hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
# Insights-> from the below graphs we can say that the most of the buyers are of age group between 26-35 and the are mostly female. 
```

```python
# Here we want to find total amount spent by each age-group
ax=df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(data=ax, x="Age Group", y= "Amount",palette=["red","green","blue","yellow","pink","black","skyblue"])

#  Insights-> From below and graphs we can see that most of the buyers are of age group between 26-35 years female
```

```python
# find total number of orders from top 10 states
ax=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
# set() func ->This is used to set the aesthetic parameters for Seaborn plots, like style, size, and other configurations.
# rc Stands for "runtime configuration". It allows you to customize Matplotlib parameters
# figure.figsize:This is a Matplotlib parameter that defines the width and height of the figure in inches.
sns.set(rc={"figure.figsize":(16,5)})
sns.barplot(data=ax, x = "State", y="Orders")

```

```python
# Total amount/sales from top 10 states
ax=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,5)})
sns.barplot(data=ax, x = "State", y="Amount")

# Insights-> from above and below graph we can say that most of the orders are from Uttar Pradesh, Maharashtra, and Karnataka respectively but total sales/amount is from UP, Karnataka and Maharashtra.
```

```python
#  Marital Status, if zero means they are married otherwise not married
sns.set(rc={"figure.figsize":(5,5)})
ax=sns.countplot(data=df, x="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
# Insights-> most of the married person are shopping more as compared to unmarried person
```

```python
ax=df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={"figure.figsize":(5,5)})
sns.barplot(data=ax, x = "Marital_Status", y="Amount", hue="Gender")
# Insights-> From above and below graph we can say that most of the buyers are married(and also female) and they have high purchasing power. 
```

```python
# Here we want to find the total count of the occupations 
sns.set({"figure.figsize":(20,5)})
ax=sns.countplot(data=df, x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  Most note-> Below is the code for sorted graph by count(total count by each occupation)
sns.set({"figure.figsize": (20, 5)})

# Sort the DataFrame by the count of "Occupation" in descending order
sorted_df = df['Occupation'].value_counts().reset_index().head(10)
# reset_index():Converts the resulting Series into a DataFrame with two columns: index (unique occupations) and their Count.
sorted_df.columns = ['Occupation', 'Count']
# sorted_df.columns :-> Renames the columns to "Occupation" and "Count" for clarity.

# Plot the sorted bar chart
ax = sns.barplot(data=sorted_df, x="Occupation", y="Count")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  Find total amount spent by each occupations
sns.set({"figure.figsize": (20, 5)})
sales_df=df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.barplot(data=sales_df, x="Occupation", y= "Amount")
#  Insights:-> from below graphs we can say that most of the buyers are working in It, Healthcare, and Aviation sector
```

```python
#  Now coming to product category
sns.set(rc={"figure.figsize":(25,5)})
ax=sns.countplot(data=df, x="Product_Category")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# find those category which has spent highest amount.
ax=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.barplot(data=ax, x="Product_Category", y="Amount")
# Insights:-> from below graph we can see that most of the sold products are from food ,Footwear and Electronics category.
```

```python
# find top selling products based on product_id
ax=df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
sns.barplot(data=ax, x="Product_ID", y="Orders")
```

```python
# above question Solved by using nlargest() func not use head(func
 
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
```
